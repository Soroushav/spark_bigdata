```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('scenario_1').getOrCreate()
```

## Load Dataset

```python
df = spark.read.csv(path='datasets/transactions_dataset.csv', header=True, inferSchema=True)
```

## Filter transactions

```python
df.filter('amount > 100').groupBy('customer_id').sum('amount').show()
```
